```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Ex 1

```python
data = pd.read_csv('data_square.csv.zip')
data.head()
```

```python
mat = np.array(data, 'float')
mat
```

```python
N = data.shape[0]
x = mat[:, 0]
y = mat[:, 1]
```

```python
#calculate coefficents of polynomial function
coefs = np.polyfit(x, y, 2)
coefs
#calculate coefficients of polynomial visualization of x
polyline = np.linspace(x[0], x[-1], len(x))
polyline
model = np.poly1d(coefs) #quadratic func
print(model)
plt.scatter(x, y)
plt.xlabel('mét vuông')
plt.ylabel('giá')
plt.plot(polyline, model(polyline), c='r') #model(polyline) -> thay số vào func model
```

## Ex 2a: fit đa thức bậc 0, 1, 3, 6, 9, vẽ mô hình và nhận xét mô hình

```python
def function(x):
    return np.sin(2 * np.pi * x)

def create_data(sample_size, std):
    x = np.linspace(0, 1, sample_size)
    t = function(x) +np.random.normal(scale = std, size = x.shape)
    return x, t     
```

```python
x, t = create_data(10, 0.25)
coefs = np.polyfit(x, t, 1)
print(coefs)
polyline = np.linspace(0, 1, 100)
t_line = function(polyline)
plt.scatter(x, t)
plt.xlabel('x')
plt.ylabel('t')
plt.plot(polyline, t_line, c = 'g')
```

```python
## bậc 1 -> underfitted
coefs = np.polyfit(x, t, 1)
polyline = np.linspace(0, 1, 100)
model = np.poly1d(coefs) 
print(model)
plt.scatter(x, t)
plt.xlabel('x')
plt.ylabel('t')
plt.plot(polyline, t_line, c = 'g')
plt.plot(polyline, model(polyline), c='r')
```

```python
## bậc 0 -> underfitted
coefs = np.polyfit(x, t, 0)
polyline = np.linspace(0, 1, 100)
model = np.poly1d(coefs) 
print(model)
plt.scatter(x, t)
plt.xlabel('x')
plt.ylabel('t')
plt.plot(polyline, t_line, c = 'g')
plt.plot(polyline, model(polyline), c ='r')
```

```python
## bậc 3 -> goodfitted
coefs = np.polyfit(x, t, 3)
polyline = np.linspace(0, 1, 100)
model = np.poly1d(coefs) 
print(model)
plt.scatter(x, t)
plt.xlabel('x')
plt.ylabel('t')
plt.plot(polyline, t_line, c = 'g')
plt.plot(polyline, model(polyline), c ='r')
```

```python
## bậc 6 -> goodfitted
coefs = np.polyfit(x, t, 6)
polyline = np.linspace(0, 1, 100)
model = np.poly1d(coefs) 
print(model)
plt.scatter(x, t)
plt.xlabel('x')
plt.ylabel('t')
plt.plot(polyline, t_line, c = 'g')
plt.plot(polyline, model(polyline), c ='r')
```

```python
## bậc 9 -> overfitted
coefs = np.polyfit(x, t, 9)
polyline = np.linspace(0, 1, 100)
model = np.poly1d(coefs) 
print(model)
plt.scatter(x, t)
plt.xlabel('x')
plt.ylabel('t')
plt.plot(polyline, t_line, c = 'g')
plt.plot(polyline, model(polyline), c ='r')
```

## thêm 15, 100 điểm dữ liệu cho đa thức bậc 9, nhận xét mức độ overfitting

```python
x, t = create_data(15, 0.25)
coefs = np.polyfit(x, t, 9)
print(coefs)
polyline = np.linspace(0, 1, 100)
model = np.poly1d(coefs) 
print(model)
plt.scatter(x, t)
plt.xlabel('x')
plt.ylabel('t')
plt.plot(polyline, t_line, c = 'g')
plt.plot(polyline, model(polyline), c ='r')
```

```python
x, t = create_data(100, 0.25)
coefs = np.polyfit(x, t, 9)
print(coefs)
polyline = np.linspace(0, 1, 100)
model = np.poly1d(np.squeeze(coefs)) 
print(model)
plt.scatter(x, t)
plt.xlabel('x')
plt.ylabel('t')
plt.plot(polyline, t_line, c = 'g')
plt.plot(polyline, model(polyline), c ='r')
```
